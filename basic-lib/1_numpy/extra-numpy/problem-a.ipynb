FILTERING

```python
data.loc[(data['Type 1'] == 'Grass') & (data['Type 2'] == 'Poison')].head(3)
```

```python
data.loc[(data['Type 1'] == 'Grass') | (data['Type 2'] == 'Poison')][0::10].loc[data['HP'] > 50].head(3) 
# Select rows where Type 1 is Grass or Type 2 is Poison, then select every 10th row, then select rows where HP is greater than 50, then display the first 3 rows
```

```python
import re
data.loc[data['Name'].str.contains('Mega')].head(3) # Select rows where Name contains "Mega" and display the first 3 rows
```

```python
data.loc[data['Type 1'].str.contains('Fire|grass', regex=True)].head(3)
```

```python
data.loc[data['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)].head(3)
# Using regex to select rows where Type 1 contains "fire" or "grass" and is case insensitive
```

```python
data.loc[data['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)].head(3)
# Using regex to select rows where Name contains "pi" and is case insensitive
```

```python
data.loc[data['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)].head(3)
# Using regex to select rows where Name starts with "pi" and is case insensitive
```

```python
data.loc[data['Name'].str.contains('^pi.*', flags=re.I, regex=True)].head(3)
# Using regex to select rows where Name starts with "pi" and is case insensitive
```
