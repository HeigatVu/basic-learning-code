```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
bg = cv2.imread('data/background.png', 1)
bg = cv2.resize(bg, (640, 480))

img = cv2.imread('data/StillImage.png', 1)
img = cv2.resize(img, (640, 480))

# Calculate the difference between the background and the image
difference = cv2.absdiff(bg, img) # Pixel similarity will be near 0 if the pixel is the same, otherwise it will be near 255
_, difference_binary = cv2.threshold(difference, 15, 255, 
                                     cv2.THRESH_BINARY) # difference > 15 will be 255, otherwise it will be 0

new_bg = cv2.imread('data/FakeBackground.png')
new_bg = cv2.resize(new_bg, (640, 480))

# Replace the background with the new background
output = np.where(difference_binary==0, new_bg, img) # If the pixel is 0, replace it with the new background, otherwise it will be the original image
cv2.imwrite('output_111.png', output)

# Show the result
plt.imshow(output)
plt.axis('off')
plt.show()
```
