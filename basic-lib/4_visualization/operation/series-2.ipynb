```python
import pandas as pd
import numpy as np

center1 = pd.Series([12, 23, 31, 11, 9], 
                    index=list(['C++', 'Golang', 'Java', 'Python', 'Swift']), 
                    name='num_registered')
center1
```

```python
center2 = pd.Series([42, 18, 44, 49, 27], 
                    index=list(['C++', 'Golang', 'Java', 'Python', 'Swift']), 
                    name='num_registered')
center2
```

```python
total = center1 + center2
total
```
