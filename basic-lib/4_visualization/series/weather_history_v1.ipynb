```python
import numpy as np
import pandas as pd

df = pd.read_csv('weatherHistory_v1.csv')
data_series = df['Temperature (C)']
#data_series = pd.Series(data=df.iloc[:,0])
data_series.info()
```

```python
data_series.plot.box() # Plot the box plot of the data series
```

```python
data_series.plot() # Plot the data series
```

```python
data_series[:100].plot() # Plot the first 100 elements of the data series
```

```python
data_series[:15] # Display the first 15 elements of the series
```

```python
data_filled = data_series.ffill() # Forward fill the missing values
data_filled[:15]
```

```python
data_filled[:100].plot()
```

```python
data_smooth = data_filled.rolling(window=5).mean() # Smooth the data using a rolling window of 5
data_smooth[:100].plot()
```

```python
data_smooth = data_filled.ewm(com=2).mean() # Exponentially weighted moving average
data_smooth[:100].plot()
```
