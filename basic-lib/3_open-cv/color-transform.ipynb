```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([1, 2.5])
print(data.dtype)

data = data.astype(np.uint8)
print(data.dtype)
```

```python
data = np.array([0, 200])
data = data.astype(np.uint8)
print(data)

data = data + 100
print(data) # result [100, 44] because 200 + 100 = 300, but the max value of uint8 is 255, so the result is 44 (255 % 200 = 44)
```

IMAGE TRANSFORM

Increasing bright

```python
# Read the image
image = cv2.imread('image1.png', 1)
image = image + 100 # Increase the brightness of the image by 100

# Convert and save
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Read the image
image = cv2.imread('image1.png', 1)
image = np.where(image + 100 > 255, 
                 255, image + 100) # if the value is greater than 255, set it to 255, but when int + 100 > 255, it will be 44

# Convert and save
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# So we need to change into float
# Read the image
image = cv2.imread('image1.png', 1)
image = np.where(image.astype(float) + 100 > 255, 
                 255, image + 100) # This method can save value without auto convert another value as int

# Convert and save
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Other method to change the brightness of the image
# Read the image
image = cv2.imread('image1.png', 1)
image = image.astype(float)

image = image + 100
image = np.clip(image, 0, 255)
image = image.astype(np.uint8)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Read the image
image = cv2.imread('image1.png', 1)
image = image.astype(float)

image = image + 100
image = np.where(image>255, 255, image)
image = image.astype(np.uint8)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```
