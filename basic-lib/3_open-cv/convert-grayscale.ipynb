```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data = np.arange(9).reshape((3,3))
print(data)
```

```python
result0 = np.apply_along_axis(np.max, arr=data, axis=0)
print(result0)
```

```python
result1 = np.apply_along_axis(np.max, arr=data, axis=1)
print(result1)
```

```python
def my_sum(vector):
    return vector.sum()

data = np.arange(9).reshape((3,3))
print(data)

result0 = np.apply_along_axis(my_sum, 
                              arr=data, axis=0)
print(result0)

result1 = np.apply_along_axis(my_sum, 
                              arr=data, axis=1)
print(result1)
```

Change each RGB value

```python
# Read the image
image = cv2.imread('image1.png', 1)

def color2grayscale(vector):
    return vector[0]*0.11 + vector[1]*0.59 + vector[2]*0.3

# Convert 
grayscale_image = np.apply_along_axis(color2grayscale, axis=2, arr=image)

plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def color2grayscale(vector):
    return vector[0]*0.33 + vector[1]*0.33 + vector[2]*0.34

# Convert to grayscale
grayscale_image = np.apply_along_axis(color2grayscale, 
                                      axis=2, 
                                      arr=image)

plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def color2grayscale(vector):
    result = vector.mean()
    result = result.astype(np.uint8)
    return result

# Convert 
grayscale_image = np.apply_along_axis(color2grayscale, 
                                      axis=2, 
                                      arr=image)

plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('image1.png', 1)
transform = np.array([0.072, 0.213, 0.715])

output = np.dot(img, transform)
plt.imshow(output, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('image1.png', 1)
transform = np.array([0.072, 0.213, 0.715])

output = np.einsum('ijk,k->ij', 
                   img, 
                   transform)

plt.imshow(output, cmap='gray')
plt.axis('off')
plt.show()
```
